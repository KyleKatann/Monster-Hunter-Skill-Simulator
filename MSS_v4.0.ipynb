```python
import numpy as np
import pandas as pd
import itertools as it

DFarmer = pd.read_table(r'..\MSS\MHWarmer.txt', encoding="cp932", header=None, sep=',')
DFskill = pd.read_table(r'..\MSS\MHWskmax.txt', encoding="cp932", header=None, sep=',')
DFarmer.drop(list(range(9,15)),axis=1,inplace=True)

SkillSlotGroup1 = DFarmer.groupby(3)
SkillSlotGroup2 = DFarmer.groupby(5)
SkillSlotGroup3 = DFarmer.groupby(7)

DFarmerArray = np.empty([1,12])

for skill in InputSkill:
    try:
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup1.get_group(skill) ) ]
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup2.get_group(skill) ) ]
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup3.get_group(skill) ) ]
    except:
        continue

DFarmerUpdated = pd.DataFrame( np.delete(DFarmerArray,0,0) ) #ゴミ処理

PartsGroup = DFarmerUpdated.groupby(1)
#print([ np.array(PartsGroup.get_group(part)) for part in Parts])
a=np.array(PartsGroup.get_group("頭"))
b=np.array(PartsGroup.get_group("胴"))
c=np.array(PartsGroup.get_group("腕"))
d=np.array(PartsGroup.get_group("腰"))
e=np.array(PartsGroup.get_group("脚"))
list(it.product(a,b,c,d,e))
```
