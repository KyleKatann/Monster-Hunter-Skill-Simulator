```python
print("スキル番号を半角数字で入力してください、なにも入力せずにEnterで終了。")
InputSkill=[]
InputSkillNum=[]
while True:
    Input=input()
    if Input=="":
        print(InputSkill)
        break
    try:
        num=int(Input)
        temp=Input_To_Skill[num].rstrip(" ")
        if temp in InputSkill:
            print("そのスキルは入力済みです。")
            continue
        print(temp)
        InputSkillNum.append(num)
        InputSkill.append(temp)
    except:
        print("半角数字で入力してください。")
#InputSkill = np.array(InputSkill)
```

```python
import numpy as np
import pandas as pd
import itertools as it
import collections as co

DFarmer = pd.read_table(r'..\MSS\MHWarmer.txt', encoding="cp932", header=None, sep=',')
DFskill = pd.read_table(r'..\MSS\MHWskmax.txt', encoding="cp932", header=None, sep=',')
DFarmer.drop(list(range(9,15)),axis=1,inplace=True)

SkillSlotGroup1 = DFarmer.groupby(3)
SkillSlotGroup2 = DFarmer.groupby(5)
SkillSlotGroup3 = DFarmer.groupby(7)

DFarmerArray = np.empty([1,12])

for skill in InputSkill:
    try:
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup1.get_group(skill) ) ]
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup2.get_group(skill) ) ]
        DFarmerArray = np.r_[ DFarmerArray , np.array( SkillSlotGroup3.get_group(skill) ) ]
    except:
        continue

        

DFarmerUpdated = pd.DataFrame( np.delete(DFarmerArray,0,0) ).replace(Skill_To_Num).replace(NABC_To_Slot) #ゴミ処理と置換
#print(DFarmerUpdated)



PartsGroup = DFarmerUpdated.groupby(1)
#print([ np.array(PartsGroup.get_group(part)) for part in Parts])
a=[list(i) for i in np.array(PartsGroup.get_group("頭"))]
b=[list(i) for i in np.array(PartsGroup.get_group("胴"))]
c=[list(i) for i in np.array(PartsGroup.get_group("腕"))]
d=[list(i) for i in np.array(PartsGroup.get_group("腰"))]
e=[list(i) for i in np.array(PartsGroup.get_group("脚"))]

OneSet=np.array(list(it.product(a,b,c,d,e)))


SkillLimitArray=np.array([Skill_Limit[Num] for Num in InputSkillNum])
#print(SkillLimitArray)
LimitBool =[]
SkillScore=[]

for oneset in OneSet[:,:,3:9].astype(int):
    dic=co.defaultdict(int)
    for(k,v)in oneset.reshape(-1,2):
        dic[k]+=v
    Scores=[dic[Num] for Num in InputSkillNum]
    LimitBool.append( all( np.array( Scores ) <= SkillLimitArray ))
    SkillScore.append(Scores)

#print(LimitBoolArray)
SlotScore=np.sum(np.sum(OneSet[:,:,9:12].astype(int),axis=1),axis=1)
LimitedRowScore = np.array(list(enumerate( np.sum( np.r_["1",np.array(SkillScore) , SlotScore.reshape(-1,1)] ,axis=1 ) )))[LimitBool]
SortedRowScore  = np.array([[k,v]for k,v in sorted( LimitedRowScore ,key=lambda x:x[1] )[::-1] ])


for i in range(100):
    print(OneSet[SortedRowScore[:,0][i]][:,0:2].reshape(10),SkillScore[SortedRowScore[:,0][i]],SortedRowScore[:,1][i])
```
