```python
#Lは5個一組を指定、その五個の中にどれだけスキルがあるか

def SKSUMfive(Linputskill,L):
    Lskillcountsum=[]
    for sss in range(len(Linputskill)):
        Lskillcount=[]
        for ttt in range(5):
            for rrr in range(len(L[ttt])):
                kkk=set([str(Linputskill[sss]),"Kyle"])
                lll=set([str(L[ttt][rrr]),"Katarn"])
                if kkk & lll != set():
                    Lskillcount.append(list(kkk&lll))
        Lskillcountsum.append(len(Lskillcount))
    
    return Lskillcountsum
Linputskill.sort()

print(len(MHWSKMAXall))
for xxx in range(len(MHWSKMAXall)):
    print(SKSUMfive(Linputskill,MHWSKMAXall[xxx]))

#L=HWSKMAXall[0]
#SKSUMfive(Linputskill,L)
```
